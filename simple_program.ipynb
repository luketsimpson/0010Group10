# Fun Number Facts Program
### A Simple Python Program

This program takes your name and a favorite number, then tells you fun facts about it!

```python
# Get user inputs
name = input("What's your name? ")
favorite_number = int(input("What's your favorite number? "))

# Do Somethingâ„¢ with the number!
print(f"\nHello, {name}! Let's explore your number {favorite_number}...\n")

# Calculate some fun facts
doubled = favorite_number * 2
squared = favorite_number ** 2
is_even = "even" if favorite_number % 2 == 0 else "odd"

# Display the results
print(f"âœ¨ Fun Facts About {favorite_number} âœ¨")
print(f"- Your number doubled is: {doubled}")
print(f"- Your number squared is: {squared}")
print(f"- Your number is {is_even}!")

# Add a special message
if favorite_number < 10:
    print(f"- Wow, {name}! That's a nice single-digit number!")
elif favorite_number == 42:
    print(f"- {name}, you picked the Answer to Life, the Universe, and Everything!")
elif favorite_number > 100:
    print(f"- {name}, you're thinking BIG with that number!")
else:
    print(f"- {name}, that's a great choice!")

print(f"\nThanks for playing, {name}! ðŸŽ‰")
```

### How it works:
1. Takes your **name** as input
2. Takes your **favorite number** as input
3. Does Somethingâ„¢ (calculates fun math facts about your number!)
